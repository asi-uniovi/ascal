```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV
df = pd.read_csv("hungarianperf-m.csv")

# Clean column names (remove leading/trailing spaces)
df.columns = [col.strip() for col in df.columns]

# Extract columns
M = df['M']
avg_vals = df['avg']
max_vals = df['max']

# Compute log-log linear fit for avg_vals
log_M = np.log10(M)
log_avg = np.log10(avg_vals)

# Fit line: log(avg) = a * log(M) + b
a, b = np.polyfit(log_M, log_avg, deg=1)

# Generate fitted line in original scale
fitted_avg_vals = 10**(a * log_M + b)

# Create single plot
plt.figure(figsize=(8, 6))

# Plot avg and max
plt.plot(M, avg_vals, marker='o', label='Avg', color='blue')
plt.plot(M, max_vals, marker='o', label='Max', color='red')

# Plot best-fit line
plt.plot(M, fitted_avg_vals, linestyle='--', color='green', label=f'Avg. best-fit (y ∝ M^{a:.2f})')

# Use log-log scale
plt.xscale('log')
plt.yscale('log')

# Labels and title
plt.xlabel('M (Number of nodes)')
plt.ylabel('Calculation time (s)')
plt.title('Hungarian Solver Performance (log-log)')
plt.grid(True, which="both", ls="--")
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV file
df = pd.read_csv("hungarianperf-cc.csv")  # Replace with your actual file name

# Clean column names
df.columns = [col.strip() for col in df.columns]

# X-axis
cc = df['CC']

# M values and corresponding column names
m_values = [10, 20, 40, 80]
avg_columns = [f'avg({m})' for m in m_values]

# Create plot
plt.figure(figsize=(8, 6))

# Plot all avg(M) curves with updated labels
for m, col in zip(m_values, avg_columns):
    plt.plot(cc, df[col], marker='o', label=f'M = {m} nodes')

# Fit best-fit line (log-log) for M = 80
x = np.array(cc)
y = np.array(df['avg(80)'])

# Remove zeros to avoid log(0)
mask = (x > 0) & (y > 0)
log_x = np.log10(x[mask])
log_y = np.log10(y[mask])

# Linear regression in log-log space
coeffs = np.polyfit(log_x, log_y, deg=1)
slope, intercept = coeffs

# Compute fitted line in original scale
fitted_y = 10**intercept * x**slope
a = 10**intercept

# Format proportionality equation for label
fit_label = f'Best-fit (M = 80): y ∝ M^{slope:.2f}'

# Plot best-fit line
plt.plot(x, fitted_y, linestyle='--', color='black', label=fit_label)

# Set log-log scale
plt.xscale('log')
plt.yscale('log')

# Add labels, title, legend, and grid
plt.xlabel('Number of container classes')
plt.ylabel('Average time (s)')
plt.title('Average Calculation Time for Different M Values (log-log)')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```
