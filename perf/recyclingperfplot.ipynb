# Performance of node recycling algorithms 

This notebook checks the performance of different solvers for the problem of node recycling.
It proves that the Hungarian Algorithm is the best event in terms of calculation time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("performance.csv")

# Clean column names
df.columns = [col.strip() for col in df.columns]

# X-axis
M = df['M']

# Columns to plot and their new labels
avg_columns = ['avg(g)', 'avg(1)', 'avg(2)', 'avg(4)', 'avg(8)']
label_map = {
    'avg(g)': 'avg(greedy)',
    'avg(1)': 'avg(hungarian)',
    'avg(2)': 'avg(hungarian, 2 parts)',
    'avg(4)': 'avg(hungarian, 4 parts)',
    'avg(8)': 'avg(hungarian, 8 parts)'
}
colors = ['black', 'blue', 'green', 'orange', 'red']

# Plot
plt.figure(figsize=(10, 6))

for col, color in zip(avg_columns, colors):
    plt.plot(M, df[col], marker='o', label=label_map[col], color=color)

# Log-log axes
plt.xscale('log')
plt.yscale('log')

# Labels and formatting
plt.xlabel('M (Number of nodes)')
plt.ylabel('Calculation time (s)')
plt.title('Recycling Solver Performance: Average Times (log-log)')
plt.grid(True, which="both", linestyle="--")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot relative loss optimality of different recycling heuristic algorithms.
# This script assumes that the performance data has been saved in a CSV file named "recycling.csv"
# with columns 'M', 'avg(g)', 'max(g)', 'avg(2)', 'max(2)', 'avg(4)', 'max(4)', 'avg(8)', 'max(8)'
# representing the number of nodes and the corresponding relative losses for different algorithms.
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("recycling.csv")

# Clean column names
df.columns = [col.strip() for col in df.columns]

# X-axis values
M = df['M']

# Columns to plot and their custom labels
avg_columns = ['avg(g)', 'avg(2)', 'avg(4)', 'avg(8)']
label_map = {
    'avg(g)': 'avg(greedy)',
    'avg(2)': 'avg(hungarian, 2 parts)',
    'avg(4)': 'avg(hungarian, 4 parts)',
    'avg(8)': 'avg(hungarian, 8 parts)',
}
colors = ['black', 'green', 'orange', 'red']

# Plot
plt.figure(figsize=(10, 6))

for col, color in zip(avg_columns, colors):
    plt.plot(M, df[col], marker='o', label=label_map[col], color=color)

# Log-log axes
plt.xscale('log')
plt.yscale('log')

# Labels and formatting
plt.xlabel('M (Number of nodes)')
plt.ylabel('Relative Loss Optimality')
plt.title('NOde recycling optimality loss (log-log)')
plt.grid(True, which="both", linestyle="--")
plt.legend()
plt.tight_layout()
plt.show()
```
